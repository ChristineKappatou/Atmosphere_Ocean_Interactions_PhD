```python
from __future__ import annotations

# load useful packages
import xarray as xr
import numpy as np
xr.set_options(display_style='html')
import xesmf as xe
from matplotlib.gridspec import GridSpec
from scipy import integrate
import matplotlib.pyplot as plt
import general_util_funcs as guf
%matplotlib inline
```

```python
def get_dx(_ds):
    """
    This function retrieves the files, latitude, and longitude information
    in all the directories within the chosen dates.
    """
    Earth_rad=6.371e6 # Earth Radius in meters

    lon = _ds['lon'].data   
    #lon[lon > 180.] -= 360 # shift longitude from 0-360˚ to -180-180˚
    lat = _ds['lat'].data

   
    dlon = np.abs(lon[1]-lon[0])
    dlat = np.abs(lat[1]-lat[0])

    lon2d,lat2d = np.meshgrid(lon,lat)
            
    dx = Earth_rad*np.cos(lat2d*np.pi/180)*dlon*np.pi/180
    # Variables to return
    return dx
```

```python
def dpres(_ds):
    """
    Calculates pressure layer thicknesses of a hybrid coordinate system
    and returns a DataArray.
    """
    ps = _ds["PS"].squeeze()  # shape: (time, lat, lon)
    hyai = _ds["hyai"].squeeze()  # shape: (nlev+1)
    hybi = _ds["hybi"].squeeze()  # shape: (nlev+1)
    p0 = _ds["P0"].squeeze()  # scalar or 1D

    # Compute interface pressures
    #p_interface = p0 * hyai + ps[:, None, :] * hybi
    # expand dims for broadcasting
    
    # p0_exp = p0.expand_dims({'ilev': hyai.ilev, 'lat': ps.lat, 'lon': ps.lon}).transpose('time', 'ilev', 'lat', 'lon')
    # hyai_exp = hyai.expand_dims({'time': ps.time, 'lat': ps.lat, 'lon': ps.lon}).transpose('time', 'ilev', 'lat', 'lon')
    # hybi_exp = hybi.expand_dims({'time': ps.time, 'lat': ps.lat, 'lon': ps.lon}).transpose('time', 'ilev', 'lat', 'lon')
    # ps_exp = ps.expand_dims({'ilev': hyai.ilev}).transpose('time', 'ilev', 'lat', 'lon')

    p0_exp = p0.expand_dims({'ilev': hyai.ilev, 'lat': ps.lat, 'lon': ps.lon}).transpose( 'ilev', 'lat', 'lon')
    hyai_exp = hyai.expand_dims({'lat': ps.lat, 'lon': ps.lon}).transpose('ilev', 'lat', 'lon')
    hybi_exp = hybi.expand_dims({'lat': ps.lat, 'lon': ps.lon}).transpose( 'ilev', 'lat', 'lon')
    ps_exp = ps.expand_dims({'ilev': hyai.ilev}).transpose('ilev', 'lat', 'lon')

    # Now calculate pressure at interfaces:
    p_interface = p0_exp * hyai_exp + ps_exp * hybi_exp

    # Then layer thicknesses = p_interface[:, i+1, :, :] - p_interface[:, i, :, :]
    pdiff = p_interface.isel(ilev=slice(1, None)).values - p_interface.isel(ilev=slice(0, -1)).values

    return xr.DataArray(
        pdiff,
        dims=( "lev", "lat", "lon"),
        coords={
           # "time": _ds.time,
            "lev": _ds.lev,
            "lat": _ds.lat,
            "lon": _ds.lon
        },
        name="dpres"
    )
```

```python
def dpres_zm(_ds):
    """
    Calculates pressure layer thicknesses of a hybrid coordinate system
    and returns a DataArray.
    """
    ps = _ds["PS"].mean(dim="lon").squeeze()  # shape: (time, lat, lon)
    hyam = _ds["hyam"].squeeze()  # shape: (nlev+1)
    hybm = _ds["hybm"].squeeze()  # shape: (nlev+1)
    p0 = _ds["P0"].squeeze()  # scalar or 1D

    # Compute interface pressures
    #p_interface = p0 * hyam + ps[:, None, :] * hybm
    # expand dims for broadcasting
    
    # p0_exp = p0.expand_dims({'ilev': hyai.ilev, 'lat': ps.lat, 'lon': ps.lon}).transpose('time', 'ilev', 'lat', 'lon')
    # hyai_exp = hyai.expand_dims({'time': ps.time, 'lat': ps.lat, 'lon': ps.lon}).transpose('time', 'ilev', 'lat', 'lon')
    # hybi_exp = hybi.expand_dims({'time': ps.time, 'lat': ps.lat, 'lon': ps.lon}).transpose('time', 'ilev', 'lat', 'lon')
    # ps_exp = ps.expand_dims({'ilev': hyai.ilev}).transpose('time', 'ilev', 'lat', 'lon')

    p0_exp = p0.expand_dims({'lev': hyam.lev,  'lat': ps.lat}).transpose( 'lev', 'lat')
    hyam_exp = hyam.expand_dims({'lat': ps.lat}).transpose('lev', 'lat')
    hybm_exp = hybm.expand_dims({'lat': ps.lat}).transpose( 'lev', 'lat')
    ps_exp = ps.expand_dims({'lev': hyam.lev}).transpose('lev', 'lat')

    # Calculate pressure at interfaces
    p_interface = p0_exp * hyam_exp + ps_exp * hybm_exp  # shape (lev, lat)

    # Create an upper boundary zero level to prepend at lev=0
    zero_level = xr.zeros_like(p_interface.isel(lev=0))
    zero_level = zero_level.assign_coords(lev=0)

    # Concatenate zero_level on top of p_interface along lev dimension
    p_interface_extended = xr.concat([zero_level, p_interface], dim='lev')
    # Now compute layer thicknesses as difference between consecutive interfaces
    pdiff = p_interface_extended.isel(lev=slice(1, None)).values - p_interface_extended.isel(lev=slice(0, -1)).values
    return xr.DataArray(
        pdiff.T,
        dims=( "ilev", "lat"),
        coords={
           # "time": _ds.time,
            "ilev": _ds.ilev.isel(ilev=slice(0, -1)),
            "lat": _ds.lat,
        },
        name="dpres"
    )
```

```python
def calc_atm_MOC(_ds, var="Vzm"):
    """
    Calculate cumulative vertical integral of meridional velocity 
    (MOC streamfunction) from top to bottom.
    """
    Earth_rad = 6.371e6  # meters
    g = 9.81 # m/s2
    lat_rad = np.deg2rad(_ds.lat)
    coslat = np.cos(lat_rad)

    Vzm = _ds[var].squeeze()  # (time, lev, lat, lon or lat)
    dp = dpres_zm(_ds)  # (time, lev, lat, lon)
    # Multiply by delta p (Pa)
    vdp = Vzm.isel(ilev=slice(0, -1)) * dp
    coslat = coslat.expand_dims({'ilev': vdp.ilev}).transpose('ilev', 'lat')
    # Cumulative sum along lev dimension from top of atmosphere downward
    # Assuming lev is ordered from top (low p) to bottom (high p)
    cumint = vdp.cumsum(dim="ilev")  
    # Apply scaling: convert Pa*m/s -> Sv (1e-15 m³/s)
    result = 1e-10 * 2 * np.pi * Earth_rad /g * coslat*cumint

    return result
```

```python
def aht_fig(_ds):
    varlist_names =  ["total", "latent", "dry_static", "kinetic", "potential", "internal"]
    varlist = ["TOTEnTrans", "LATEnTrans", "POTEnTrans", "KINEnTrans", "INTEnTrans",]
    fig, axs = plt.subplots( 2, 3, layout="constrained", figsize=(9, 6))
    axs = axs.flatten()
    k=0
    for var in varlist: 
        ax = axs[k]
        da = _ds[var].sum(dim="lev")#.mean(dim="time")
        dx = get_dx(da)
        da = 1e-15*(dx*da).sum(dim = "lon" ).squeeze()
        p = ax.plot(da.lat, da, label = varlist_names[k] , color="tab:purple")
        ax.grid()
        ax.legend()
        k=k+1
```

```python
def moc_fig(_ds):
    da = calc_atm_MOC(_ds, var="Vzm")
    # Assuming da has dims ('lev', 'lat')
    ilev = da['ilev'].values
    lat = da['lat'].values

    # Create meshgrid for lev and lat
    LAT, LEV = np.meshgrid(lat, ilev)

    fig, ax = plt.subplots(figsize=(8,6))

    # Plot contourf, levels can be adjusted or auto
    cf = ax.contourf(LAT, LEV, da.values, cmap='coolwarm', levels=20)
    ax.contour(LAT, LEV, da.values, colors='silver', linewidths = .5, levels=10)

    # Flip y-axis (pressure decreases upward)
    ax.invert_yaxis()
    ax.set_ylabel('Sigma level [hPa]', fontsize = 12)
    ax.set_xlabel('Latitude [degree north]', fontsize = 12)
    ax.set_title(f'Meridional overturning circulation', fontsize = 12)
    ax.set_ylim([1000,200])
    cbar = fig.colorbar(cf, ax=ax)
    cbar.set_label("MOC [kg/s]", fontsize = 14)

    plt.show()
```

```python
def zonal_wind_fig(_ds):
    da = _ds["Uzm"].squeeze()
    ilev = da['ilev'].values
    lat = da['lat'].values

    # Create meshgrid for lev and lat
    LAT, LEV = np.meshgrid(lat, ilev)

    fig, ax = plt.subplots(figsize=(8,6))

    # Plot contourf, levels can be adjusted or auto
    cf = ax.contourf(LAT, LEV, da.values, cmap='coolwarm', levels=20)
    ax.contour(LAT, LEV, da.values, colors='silver', linewidths = .5, levels=10)

    # Flip y-axis (pressure decreases upward)
    ax.invert_yaxis()

    ax.set_xlabel('Latitude')
    ax.set_ylabel('Sigma level [hPa]')
    ax.set_title(f'Zonal mean winds')
    ax.set_ylim([1000,100])
    fig.colorbar(cf, ax=ax, label=da.name)

    plt.show()
```
